**1) Print a list**

```python
a = [3, 8, 5]
print(a)
```

**2) Adding another element to list**
Add an element to the end of the list

```python
a.append(1)
print(a)
```

**Add an item at the specified index**

Use the insert() method

```python
b = [102, 318, 97, 232]
b.insert(2, 666)
print(b)
```

**3) Variable types can be mixed e.g. integers and strings**

```python
a.append("hello")
print(a)
```

```python
a.append([6, 9])
print(a)
```

**5) Deleting an item from a list**

```python
a.pop() #Deletes the last item from the list
print(a)
```

**Remove an item from list**

```python
d = [3, 8, 5, 1, 'hello', 'dinosaur']
d.pop(4)
print(d)
```

```python
d = [3, 8, 5, 1, 'hello', 'dinosaur']
del d[3]
print(d)
```

```python
g = [3, 8, 5, 1, 'hello', [6, 9]]
g.remove([6, 9])
print(g)
```

**Clear the entire list**

```python
d = [3, 8, 5, 1, 'hello', 'dinosaur']
d.clear()
print(d)
```

**7) Change the value of an index to another**

```python
a[0] = 100
print(a)
```

```python
#EX swap the first index with the last
b = ["banana", "apple", "alien"]

#Define a new variable and let it be the value of index 0
temp = b[0]

#Equate index 0 to the last index of list
b[0] = b[2]

#Now let last index equal the first variable
b[2] = temp

print(b)
```

```python
#Alternatively
b[0], b[2] = b[2], b[0]
print(b)
```

**Iterate or Loop through a list**

Print all elements in a list one by one.

Use a for loop for this.

```python
thislist = ["raspberry", "strawberry", "blackberry"]
for item in thislist:
    print(item)
```
