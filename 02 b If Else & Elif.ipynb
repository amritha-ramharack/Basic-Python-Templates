```python
#Eg 1 Elif means "if the previous conditions were not true, then try this condition"

a = 44
b = 44
if b>a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
    
```

```python
#Eg 2 The else keyword catches anything which isn't caught by the preceding conditions.

c = 300
d = 44
if d>c:
    print("d is greater than c")
elif c == d:
    print("c and d are equal")
else:
    print("c is greater than d")
```

```python
#Eg 3 If you have only one statement to execute, you can put it on the same line as the if statement.
a = 44
b = 20

if a > b: print("a is greater than b")
```

```python
#Eg 4 If you have only one statement to execute, one for if, and one for else, you can put it all on the same line:
a = 5000
b = 20
print("A") if a >b else print ("B")
```

```python
#Eg 5 multiple else statements on the same line:
x = 26
y = 25
print("X") if x > y else print("=") if x==y else print("Y")
```

```python
#Eg 6 The and keyword is a logical operator, and is used to combine conditional statements:
#Test if a is greater than b, AND if c is greater than a:
a = 5000
b = 20
c = 5500
if a > b and c > a:
    print("both conditions are true")
```

```python
#Eg 7 The or keyword is a logical operator, and is used to combine conditional statements:
#Test if a is greater than b, OR if a is greater than c:
a = 5000
b = 20
c = 5500
if a > b or a > c:
    print("at least one condition is true")
```
