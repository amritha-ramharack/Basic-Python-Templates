```python
c = 7
d = 4
if c < d:
    print("c is less than d")
else:
    print("c is definitely NOT less than d")
```

```python
e = 8
f = 8
if e < f:
    print("e is less than f")
elif e == f:
    print("e is equal to f")
else:
    print("e is not less than f")
```

```python
g = 11
h = 9
if g < h:
    print("g is less than h")
else:
    if g == h:
        print("g is equal to h")
    else:
        print("g is greater than h")
```

**Ternary**

```python
#Originally we write
age = 19
if age >= 21:
    old_enough = True
else:
    old_enough = False
print(old_enough)
```

```python
#Alternatively, we can code
age = 33
old_enough = True if age >= 21 else false
print(old_enough)
```
