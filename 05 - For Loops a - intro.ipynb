```python
b = ["banana", "apple", "robots"]
```

```python
#Eg 1 To retrieve any item from the list
print(b[2])
```

```python
#Eg 2a To print more than 1 item, especially from a long list, we ue FOR LOOPS
for element in b:
    print(element)
```

```python
#Eg 2b To print each item more than once:
for element in b:
    print(element)
    print(element)

```

```python
#Eg 2c
for element in b:
    print(element)
for element in b:
    print(element)
```

```python
#element above can be replaced with any other name
d = [20, 10, 5]
for q in d:
    print(q)
```

```python
#Eg 3a Looping through a string
for g in "banana":
    print(g)
```

```python
# Eg 3b BREAK STATEMENT to stop the loop before it has completed
#Exit the loop when x is banana
fruits =["apple", "strawberries", "pears", "banana", "cherry"]
for x in fruits:
    print(x)
    if x == "banana":
        break
```

```python
#Eg 3c Exit the loop when x is banana 
#Break comes before the print
fruits =["apple", "strawberries", "pears", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        break
    print(x)
```

```python
#Eg 3d CONTINUE STATEMENT where the current iteration of the loop is stopped
#We continue to the next 
#Don't print banana
fruits =["apple", "strawberries", "pears", "banana", "cherry"]
for x in fruits:
    if x == "banana": # == comparison operator. x has the same value as banana 
        continue
    print(x)
```

```python
#Eg 4
RANGE FUNCTION
#returns a sequence of numbers from 0
#Used to loop through a set of code a certain number of times.
for x in range(6):
    print(x)
```

```python
#Eg 5 Printing in increments (start, end, increment)
for x in range(2, 30, 3):
    print(x)
```

```python
#Calculating the sum of a list x
x = [20, 30, 50]

#initialize a variable and set it to 0
total = 0

#now, in the for loop, add each element to total
for element in x:
    total = total+element
print(total)
```

```python
#Print list when index maters
a = ["peach", "pear", "robot"]
for i in range(len(a)):
    print(a[i])
```

```python
a = ["peach", "pear", "robot"]
for i in range(len(a)):
    for j in range(i+1):
        print(a[i])
```
