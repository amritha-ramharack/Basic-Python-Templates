**Adding a key value pair**

After defining a dictionary, we can add a key value pair.

```python
e ={}
e["Chris"] = 36
#this says add a new key value pair where the key is Chris and the value is 38
e["Steve"] = 38
e["Scarlet"] = 34
print(e)
```

**Access the value that's associated with a certain key**

```python
print(e["Chris"])
```

```python
print(e["Steve"])
```

```python
print(e["Scarlet"])
```

**Change the value associated with a key**

This can be done even after we have that key value pair in the dictionary.

```python
print(e["Scarlet"])
```

```python
e["Scarlet"] = 35
```

```python
print(e["Scarlet"])
```

**Data types for dictionaries**

The values can beany data type.
Keys must be certain types. Keys are generally strings or integers.

**Iterating over key-value pairs**

```python
for key, value in e.items():
    print("key")
    print(key)
    print("value")
    print(value)
```

```python
#alternatively
for k, v in e.items():
    print(k, "=>", v )
    
```
