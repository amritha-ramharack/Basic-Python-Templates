## Use While Loops instead of For Loops if we don't know how many loops we need before hand. ##


```python
#We can add up all numbers from 1 to 5 in a for loop
total = 0
for i in range(1,5):
    total += i
print(total)
```

```python
#Eg 1 we can do the same thing in a While Loop
total2 = 0
j = 1 #While Loop doesn't initialise the index for us. We explicitly initialize it here.
while j < 5:
    total2 += j #Add j to the total2 variable
    j += 1 #Add 1 to j
print(total2)

#the While Loop block says to check to see if j is less than 5. If it's true then:
#a) add j to ttotal2
#b) then add 1 to j
    
```

```python
#Eg 5 We don't know the list's contents other than it's sorted in descending order. 
# We want to find the sum of all positive numbers
given_list = [5, 4, 4, 3, 1, -2, -3, -5]
total5 = 0
i = 0 #This gives the element number in the list

while given_list[i] > 0:
    total5 += given_list[i]
    i += 1
#This block says to first check the current element in given_list; Since i starts at 0,
# the first current element of given_list will be the first element, which is 5. Since this 
# element is greater than 0, do the following:
# a) add the current element, given_list[i] to total5
# b) then increment i by 1

print(total5)
```

```python
#Eg 6 If there were only positive numbers in the list, there will be an error.
#There will be an erro at line i += 1. For i = 4 we get 5, and when the program looks for index 5,
#there is none. So we need an additional condition to the while loop.

given_list = [5, 4, 4, 3, 1]
total5 = 0
i = 0 #This gives the element number in the list

while i < 5 and given_list[i] > 0:
    total5 += given_list[i]
    i += 1
print(total5)
```

```python
#Eg 7 Alternatively, we can code:
given_list = [5, 4, 4, 3, 1]
total5 = 0
i = 0 #This gives the element number in the list

while i < len(given_list) and given_list[i] > 0:
    total5 += given_list[i]
    i += 1
print(total5)
```
