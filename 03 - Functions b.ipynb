```python
#BMI Calculator
name1 = "YK"
ht1 = 2
weight1 = 90
name2 = "YK sis"
ht2 = 1.8
weight2 = 70
name3 = "YK bro"
ht3 = 2.5
weight3 = 160

```

```python
def bmi_calculator(name, ht, weight): #function is defined and argument is stated
    bmi = weight / (ht**2) 
    print("bmi: ")
    print(bmi)
    if bmi < 25:
        return name+" is not overweight"
    else:
        return name+" is overweight"
result1 = bmi_calculator(name1, ht1, weight1) \

result2 = bmi_calculator(name2, ht2, weight2)
result3 = bmi_calculator(name3, ht3, weight3)

print(result1)
print(result2)
print(result3)
    
```
