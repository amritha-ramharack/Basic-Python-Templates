```python
#Eg 1a Converting a range to a list
range(1, 5) #this is a range of numbers from 1 through 5, but not including 5

#we convert it to a list as follows
c = list(range(1, 5))
print(c)
```

```python
#Eg 1b Alternatively, we can write the range as follows
range(5)
d = list(range(5))
print(d)
```

```python
#Eg 1c Alternatively
print(list(range(1,9)))
```

```python
#Eg 2a Find the sum of a range

total = 0
for i in range (1, 5): 
    total = i + total  
print (total)
```

```python
#Eg 2b Alternatively

total = 0
for i in range (1, 5): 
    total += i  # += means that the new value of total should be the sum of old value of total plus i
print (total)
```

```python
#Eg 2c Alternatively

total2 = 0
for e in range(5):
    total2 += e
print (total2)
```

```python
#Eg 3 Summing specific numbers in the range/list
#Find the sum of only the multiples of 3

total3 = 0
for i in range(1,9):
    if i % 3 == 0: #Module operator
        total3 += i
print(total3)
```

```python
#Eg 4 Compute the sum of all the multiples of 3 and 5 that are less than 100
sum1 = 0
for q in range(1,100):
    if q % 3 == 0:
        sum1 += q
    elif q % 5 == 0:
        sum1 += q
print(sum1)
```
