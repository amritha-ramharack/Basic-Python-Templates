```python
#Eg 1 if we have a list of both positive and negative numbers, can we add all the positive numbers?
#We use a break statement to breakout of the loop when we see a negative number.

given_list2 = [5, 4, 4, 3, -1, -2, -3,-5]

total4 = 0
for element in given_list2:
    if element <= 0:
        break
    total4 += element
print(total4)
```

```python
#Eg 2 Break statements can be used in while loops too.

given_list2 = [5, 4, 4, 3, -1, -2, -3,-5]
total5 = 0
i = 0 #index number
while True: #this statement must always be true
    total5 += given_list2[i]
    i += 1
    if given_list2[i] <= 0:
        break
print(total5)
```

```python
#Eg 3 We have a list sorted in decending order but we don't know its contents.
#What is the sum of all the negative nmbers?

given_list3 = [7, 5, 4, 4, 3, 1, -2, -3, -5, -7]

total2 = 0
j = len(given_list3)-1 #j is the index that corresponds to the last element on the list
while given_list3[j] < 0:
    total2 += given_list3[j] #add the element to total2
    j -= 1
print(total2)
```
